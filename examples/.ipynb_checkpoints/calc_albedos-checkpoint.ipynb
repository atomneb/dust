```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import dust
import sigma_scat as ss
import scatmodels as sm
```

```python
mrn = dust.MRN_dist(0.005, 0.3, 3.5)
```

```python
sil_mie = ss.makeScatmodel('Mie','Silicate')
gra_mie = ss.makeScatmodel('Mie','Graphite')
```

```python
EGRID = np.logspace(np.log10(0.3), np.log10(8.0), 20)
```

```python
sil_kappasca = ss.Kappascat(E=EGRID, scatm=sil_mie, dist=mrn)
sil_kappaext = ss.Kappaext(E=EGRID, scatm=sil_mie, dist=mrn)
```

```python
%%time
gra_kappasca = ss.Kappascat(E=EGRID, scatm=gra_mie, dist=mrn)
gra_kappaext = ss.Kappaext(E=EGRID, scatm=gra_mie, dist=mrn)
```

```python
sil_albedo = sil_kappasca.kappa / sil_kappaext.kappa
gra_albedo = gra_kappasca.kappa / gra_kappaext.kappa
```

```python
plt.plot(EGRID, sil_albedo, 'g-')
plt.plot(EGRID, gra_albedo, 'b-')
plt.semilogx()
```

## Calculate the albedo of some PAHs

Go from 0.1 to 10 um

```python
wavel_angs = np.logspace(-1, 1, 100) * 1.e4
wavel_keV  = dust.c.hc_angs / wavel_angs # keV
```

```python
pah_model = sm.PAH('neu')
```

```python
def calc_albedo(a):
    sca = pah_model.Qsca(wavel_keV, a)
    ext = pah_model.Qext(wavel_keV, a)
    return sca/ext
```

```python
albedo100 = calc_albedo(0.01)
albedo10  = calc_albedo(0.001)
```

```python
ax = plt.subplot(111)
plt.plot(wavel_angs, albedo100, label='100 Angs radius')
plt.plot(wavel_angs, albedo10,  label='10 Angs radius')
plt.loglog()
plt.ylabel(r'albedo ($Q_{sca}/Q_{ext}$)', size=14)
plt.xlabel('Wavelength (Angs)', size=14)
ax.tick_params(labelsize=12)
plt.legend(loc='upper right', frameon=False)
#plt.savefig('/Users/lia/Dropbox/Presentations/pyastro16/PAHexample.pdf')
```
